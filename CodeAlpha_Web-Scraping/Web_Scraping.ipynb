# Task 1 - Web Scraping

- Importing BeautifulSoup and Request

```python
import requests
from bs4 import BeautifulSoup
```

- Scrapping Website : https://books.toscrape.com/

```python
a = requests.get("https://books.toscrape.com/")
a.status_code
```

- Single page imported 

```python
with open("HTML/page1.html", "w") as p:
    p.write(a.text)
```

- Read File

```python
with open("HTML/page1.html") as f:
    content = f.read()
```

```python
soup = BeautifulSoup(content, "html.parser")                                 # making soup as Var and 
                                                                             # html.parser" tells BeautifulSoup to use Python’s built‑in HTML parser.
```

- Scrapping : Title, Price, Availability & Rating

```python
articles = soup.select("article.product_pod")
# articles
items = []                                                                    # Empty List
for article in articles:
    title = article.find("h3").find("a")['title']
    price = article.select_one("p.price_color").text.split("Â£")[1]
    Available = article.select_one("p.availability").text.split()
    rating_element = article.select_one("p.star-rating")
    rating = rating_element['class'][1]
    # print(Available)
    items.append([title, Available, price, rating])
```

- Importing pandas to make DataFrame 

```python
import pandas as pd

df = pd.DataFrame(items, columns=["Title", "Available", "Price", "Rating"])
df
```

- Saved in CSV File

```python
df.to_csv("Books.csv", index = False)
```
